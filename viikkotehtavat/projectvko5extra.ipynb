```python
import numpy as np
import pandas as pd
import keras
from keras import layers


num_classes = 6
input_shape = (3,)

df = pd.read_csv("py_data.csv",header=None)

data = df.iloc[:,6:9].to_numpy()
ans = df.iloc[:,5].to_numpy()

#left, right = keras.utils.split_dataset(data, left_size=0.9, shuffle = True)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, ans, test_size=0.20, random_state=33)
#print(left)
#print(data)
#print(ans)
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
print(y_train)

#print(y_train)
print(type(y_train))
```

```python
model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Dense(10, activation="relu"),
        #layers.Flatten(),
        layers.Dense(num_classes, activation="softmax")
    ]
)

model.summary()
#model.weights
```

```python
batch_size = 100
epochs = 1000

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.1)
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```
